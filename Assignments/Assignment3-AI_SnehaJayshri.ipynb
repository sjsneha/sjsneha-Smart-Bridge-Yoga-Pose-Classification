## To build an Artificial Neural Network (ANN) model

## The whole code:

```python
#Step 1: Import Libraries

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
from keras.models import Sequential
from keras.layers import Dense


#Step 2: Load and Preprocess the Dataset

# Read the dataset
data = pd.read_csv("/Users/91957/Desktop/AI/drug200.csv")

# Split features and labels
X = data.drop('Drug', axis=1)
y = data['Drug']

# Perform label encoding on categorical variables
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Perform one-hot encoding on categorical variables
onehot_encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
X_encoded = onehot_encoder.fit_transform(X[['Sex', 'BP', 'Cholesterol']])
X_encoded = np.hstack((X_encoded, X[['Age', 'Na_to_K']].values))

# Scale the numeric features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)




#Step 3: Build the ANN Model

# Build the ANN model
model = Sequential()
model.add(Dense(units=64, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=16, activation='relu'))
model.add(Dense(units=len(label_encoder.classes_), activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1)



#Step 4: Train and Evaluate the Model

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test, verbose=1)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)



#Step 5: Make Predictions


# Example prediction
sample_data = np.array([[50, 'M', 120, 'NORMAL', 0.9]])

# Perform one-hot encoding on the categorical variables in sample_data
sample_data_encoded = onehot_encoder.transform(sample_data[:, [1, 2, 3]])
sample_data_encoded = np.hstack((sample_data_encoded, sample_data[:, [0, 4]].astype(float)))

# Scale the numerical features in sample_data
sample_data_scaled = scaler.transform(sample_data_encoded)

# Perform predictions
predicted_class = np.argmax(model.predict(sample_data_scaled), axis=1)
predicted_drug = label_encoder.inverse_transform(predicted_class)

print("Predicted Drug:", predicted_drug)
```
